```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


df = pd.read_excel('MK_Proteomics.xlsx')

#Select only numeric columns
X = df.select_dtypes(include=['number'])

#Normalize all features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=1)
labels = kmeans.fit_predict(X_scaled)

print("Cluster assignments for each sample:")
for i, label in enumerate(labels):
    print(f"Sample {i+1}: Cluster {label}")
```

```python
#PCA reduction
pca = PCA(n_components=2, random_state=42)
pcs = pca.fit_transform(X_scaled)

#Scatter plot of first two PCs colored by cluster
plt.figure()
plt.scatter(pcs[:, 0], pcs[:, 1], c=labels)
plt.xlabel('PC${}_1$')
plt.ylabel('PC2')
plt.title('PCA of MK Proteomics Data with K-Means Cluster Colors')
plt.show()
```
